### Data Preparation

### Handling Data Type

```python
df[timestamp] = df[timestamp].apply(pd.to_datetime)
```

```python
df.info()
```

### Handling Missing Value

```python
df = df.dropna()
```

```python
df.isna().sum()
```

### Data Duplicate 

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

### Handling Outliers

### Log Transformation

```python
for num in nums:
    df[num] = np.log(df[num] + 0.5)
```

```python
df[nums].describe()
```

#### Z-score

```python
from scipy import stats

print("Before removing outlier: ", len(df))

for num in nums:
    z_scores = np.abs(stats.zscore(df[num]))
    df = df[z_scores <= 3]

print("After removing outlier: ", len(df))
```

```python
df[nums].describe()
```

### Normalization & Standarization

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

for num in nums:
    df[num] = StandardScaler().fit_transform(df[num].values.reshape(len(df), 1))
```

```python
for num in nums:
    df[num] = MinMaxScaler().fit_transform(df[num].values.reshape(len(df), 1))
```
